### seaborn을 활용한 샘플 데이터셋 활용

```python
import seaborn as sns
sns.get_dataset_names()  # dataset list 확인
```

```python
# 빌트인 데이터셋 불러오기
sns.load_dataset('tips')
```

```python
# 필요한 데이터셋 샘플 찾기
for i in sns.get_dataset_names():
    tmp = sns.load_dataset(i)
    print(i + ' : ' + str(tmp.shape))
    display(tmp)
```

```python
# 사무실에서도 쓸수 있게 피클화
for i in sns.get_dataset_names():
    sns.load_dataset(i).to_pickle(i + '.pkl')
```

### 회귀분석을 위한 사이킷런 활용

```python
from sklearn.datasets import make_regression

X, y, coef = make_regression(
    n_samples=20, # 행 갯수
    n_features=3, # 열 갯수(n+1개)
    coef = True,
    bias = 10,
    random_state=0
    )

# 컬럼명이 필요하면 columns로 지정
pd.DataFrame(data=np.hstack((X, y.reshape(-1, 1))))
```
