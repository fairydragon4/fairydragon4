```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
# bar() : 세로 막대 그래프 그리기
x = np.arange(3)
plt.bar(x, [100, 400, 900],  # 표시할 값
        color = 'y',   # 그래프 색상, 각각 지정시 ['r','g','b'] 와 같이 같은 크기의 리스트로 지정
        width = 0.4,  # 그래프 너비,  디폴트는 0.8?
        align = 'edge',  # 눈금 위치, *center | edge 왼쪽 가장자리
        edgecolor = 'black',  # 그래프 테두리 색상
        linewidth = 5,  # 그래프 테두리 두께
        tick_label = ['2019','2020','2021']  # x축 레이블
       )
# plt.xticks(x, ['2020','2021','2022'])  # x축 레이블, 이렇게도 지정 가능

plt.show()
```

```python
# barh() : 가로 막대 그래프 그리기
y = np.arange(3)
plt.barh(y, [100, 400, 900],
         color = ['C2', 'C3', 'C4'],   # 그래프 색상, 각각 지정시 ['r','g','b'] 와 같이 같은 크기의 리스트로 지정
         height = 0.9,  # 그래프 높이
         align = 'center',  # 눈금 위치, *center | edge 왼쪽 가장자리
         edgecolor = 'gold',  # 그래프 테두리 색상
         linewidth = 3,  # 그래프 테두리 두께
         tick_label = ['2019','2020','2021']  # x축 레이블
        )
```

```python
# scatter(x, y) : 산점도 그리기
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
c = np.random.rand(n)

plt.scatter(x, y,  # x, y 좌표
            s = (30 * np.random.rand(n)) ** 2,  # 그래프 크기
            c = np.random.rand(n),  # 그래프 색상
            alpha = 0.5,  # 투명도 지정
            cmap = 'Spectral'  # 컬러맵에 해당하는 문자열 지정
           )

plt.colorbar()  # 컬러바 표시
plt.show()
```

```python
# 3차원 산점도 그리기
from mpl_toolkits.mplot3d import Axes3D

n = 100
xmin, xmax, ymin , ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0, 2
xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
cs = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(xs, ys, zs, c = cs, cmap = 'Greens')

plt.show()
```

```python
# hist() : 히스토그램 그리기
x = np.random.randint(0, 30, size=100)
plt.hist(x,  # 데이터 분포
#          bins = 5,  # 가로축 구간의 개수
         cumulative = False,  # 누적 히스토그램,  True | *False
         histtype = 'bar',  # 히스토그램 종류,  *bar | barstacked | step | stepfilled
         density = False,  # true일 경우 밀도함수가 되어서 합이 0이 됨,  True | * False
         alpha = 0.7
        )

plt.show()
```

```python
# pie() : 파이 차트 그리기 - 1
x = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.1, 0, 0.3]
colors = ['silver', 'gold','whitesmoke','lightgray']
plt.pie(x,  # 비중 데이터
        labels = labels,  # 라벨
        autopct = '%.1f%%',  # 데이터 레이블 표시, 표시 형식 지정
        explode = explode,  # 파이 차트에서 벗어난 정도
        colors = colors,  # 차트 색상
        startangle = 260,  # 부채꼴 시작 각도
        counterclock = False,  # 반시계 방향으로 표시, False | *True
        shadow = True,  # 그림자 표시, *False | True
       )

plt.show()
```

```python
# 도넛 차트 그리기
x = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.1, 0, 0.3]
colors = ['#ff9999', '#ffc000','#8fd9b6','#d395d0']
wedgeprops = {'width' : 0.1, 'edgecolor' : 'w', 'linewidth' : 5}  # 고리 두께 0~1, 경계 색, 경계 두께
plt.pie(x,  # 비중 데이터
        labels = labels,  # 라벨
        autopct = '%.1f%%',  # 데이터 레이블 표시, 표시 형식 지정
        colors = colors,  # 차트 색상
        wedgeprops = wedgeprops
        )

plt.show()
```

```python
# matshow() : 히트맵 그리기
# get_cmap() : 컬러맵 종류 지정
# colorbar() : 컬러바 표시
arr = np.random.standard_normal((30, 40))
cmap = plt.get_cmap('PiYG')  # 컬러맵 종류 지정, PiYG | BuGn | Greys | bwr 등
plt.matshow(arr,
            cmap = cmap  # 컬러맵 설정
           )
plt.colorbar(  # 컬러바 표시, 옵션없이 빈값 가능
    shrink = 0.8,  # 그래프 세로 대비 길이의 비, 디폴트는 1
    aspect = 10,  # 컬러바 종횡비, 디폴트는 20, 작을수록 두꺼워짐
)
plt.clim(-2, 2),  # 색상의 범위, 범위를 벗어날 경우 해당 경계값과 같은 색으로 나타남

plt.show()
```
