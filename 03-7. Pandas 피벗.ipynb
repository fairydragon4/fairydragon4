```python
import pandas as pd
import numpy as np
import seaborn as sns
import random
```

```python
# 샘플 데이터프레임 생성_1
df1 = sns.load_dataset('healthexp')
df1
```

```python
# pivot() : 피벗 테이블 생성, 해당 행렬의 값이 하나씩만 매칭되어야 함
df1.pivot(index = 'Year',  # 인덱스로 지정
          columns = 'Country',  # 컬럼으로 지정
          values = 'Life_Expectancy'  # 값으로 지정, 미지정 시 남은 모든 열 선택
          ).head()
```

```python
# 복수 value 지정
df1.pivot(index = 'Year',  # 인덱스로 지정
          columns = 'Country',  # 컬럼으로 지정
          values = ['Spending_USD', 'Life_Expectancy']  # 값으로 지정
         ).head()
```

```python
# 복수 index 지정
df1.pivot(index = ['Year', 'Country'], # 인덱스로 지정
          columns = [],
          values = 'Life_Expectancy'  # 값으로 지정
         ).head(20)
```

```python
# 샘플 데이터프레임 생성_2
df2 = sns.load_dataset('penguins')
df2
```

```python
# pivot_table() : 피벗 테이블 생성
df2.pivot_table(values = 'bill_length_mm',  # 값으로 지정
                index = 'species',  # 인덱스로 지정
                columns = 'sex',  # 컬럼으로 지정
                aggfunc = 'mean',  # 집계 방법, *mean | sum | count | min | max 
                margins = True,  # 합계행/열 표시 여부, True | *False
                margins_name = '전체'  # 합계행/열 이름, *All
        )
```

```python
# 복수 집계합수 지정
df2.pivot_table(values = 'bill_length_mm',  # 값으로 지정
                index = 'species',  # 인덱스로 지정
                columns = 'island',  # 컬럼으로 지정
                aggfunc = ['sum', 'count'],  # 집계 방법, *mean | sum | count | min | max 
                fill_value = -1  # 결측치 채워넣을 값
        )
```

```python
# 각 value에 대해 다른 집계함수 지정 가능
df2.pivot_table(values = ['bill_length_mm', 'bill_depth_mm'],  # 값으로 지정
                index = 'species',  # 인덱스로 지정
                columns = 'sex',  # 컬럼으로 지정
                aggfunc = {'bill_length_mm' : 'sum', 'bill_depth_mm' : np.mean}
        )
```

```python
# 샘플 데이터프레임 생성_3
df3 = df2.pivot_table(values = 'bill_length_mm', index = 'species', columns = 'sex', margins = True).reset_index()
df3                
```

```python
# melt() : 언피벗 
df3.melt(id_vars = 'species',  # 기준이 될 열
         value_vars = ['Female', 'Male', 'All'],  # 하위 카테고리, 미지정 시 남은 모든 열 선택
         var_name = '구분',  # 하위 카테고리 이름, *variable
         value_name = '값',  # 값 이름, *value
         ignore_index = True  # 기존 인덱스 사용 여부, *True | False
        )
```

```python
# 샘플 데이터프레임 생성_4
dict1 = {'이름' : ['Mercury','Venus','Mars','Jupiter','Saturn','Uranus','Neptune']}
dict1['분단'] = [1, 2, 1, 1, 2, 1, 2]
dict1['국어'] = random.sample(range(50,100), 7)
dict1['수학'] = random.sample(range(30,100), 7)
dict1['영어'] = random.sample(range(40,100), 7)

df4 = pd.DataFrame(dict1) 
df4
```

```python
# stack() : 컬럼을 인덱스로 보냄, Series로 바뀜
df5 = df4.stack()
df5
```

```python
# unstack() : 인덱스를 컬럼으로 보냄
df5.unstack()
```

```python
# 샘플 데이터프레임 생성_5 -> 예시 좀더 쉽게 바꿀것
interval = [0,0,0,1,1,1,2,2,3,3,3,4,4,4,5]
axis = ['x','y','z','x','y','z','x','y','z','x','y','z','x','y','z']
values = [0,0.5,1,1.5,2,2.5,3,4,3.5,3,2.5,2.1,1.5,1.7,4]
df = pd.DataFrame({'interval': interval, 'axis' : axis, 'value': values})
name1 = df.copy()
name2 = df.copy()
name3 = df.copy()
name1['name'] = 'john'
name2['name'] = 'cash'
name3['name'] = 'eric'
name2['value'] *= 10
name3['value'] *= 100
all_name_df = pd.concat([name1, name2, name3]).set_index(['name','interval', 'axis'])
all_name_df
```

```python
# 멀티인덱스의 경우 하위인덱스부터 올라감
all_name_df.unstack()
```

```python
# 여러 인덱스 지정시 파라미터를 리스트로 지정
all_name_df.unstack(['name','interval'])
```
