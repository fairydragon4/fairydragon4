```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

### 그래프 그리기 기본

```python
plt.plot([1,2,3,4,5])  # y값만 지정, x값은 0부터 자동으로 부여
plt.plot([1,2,3], [3,4,5])  # x, y값 각각 지정
plt.plot(np.arange(0,4), [5,4,3,2])  # numpy 배열 사용

plt.show()
```

```python
import seaborn as sns
anagrams = sns.load_dataset('anagrams')
plt.plot(anagrams.num1)  # series로 그래프 그리기
```

```python
# 복수 그래프 한번에 그리기
t = np.arange(0, 5, 0.2)  # 0~5를 0.2 간격으로 두는 narray
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**(1/2), 'g^')
plt.show()
```

### 선 그래프 옵션

```python
# xlabel(), ylabel() : 축 레이블 지정
plt.plot([1,2,3,4],[2,6,4,8])
plt.xlabel('X-Axis',  # 표시 텍스트
           labelpad=15,  # 축과 레이블간 간격
           fontdict={'family' : 'serif', 'color' : 'b', 'weight' : 'bold', 'size' : 14},  # 글꼴
           loc='right'  # 위치, xlabel은 left | *center | right, ylabel은 top | *center | bottom
          )
plt.ylabel('Y-Axis', labelpad = 20, fontdict={'family':'fantasy', 'color' : 'deeppink', 'weight' : 'normal', 'size' : 'xx-large'})
plt.show()
```

```python
# legend() : 범례 표시하기
plt.plot([1,2,3,4], [1,5,1,5], label = 'Price ($)')
plt.plot([1,2,3,4], [2,3,4,5], label = 'Demand (#)')
plt.plot([1,2,3,4], [5,4,3,2], label = 'example')
plt.legend(  # 범례 표시, 옵션없이 빈값 가능
    loc = (0.75, 0.3),  # 위치 지정, (0, 0) 왼쪽 아래 ~ (1,1) 오른쪽 위 | lower right 등 | *best 가장 적합한 위치
    ncol = 2,  # 한 행에 표시할 컬럼 수
    fontsize = 14,  # 폰트 크기
    facecolor = '#F0FFE0',  # 범례 배경색
    edgecolor = 'limegreen',  # 범례 테두리색
    frameon = True,  # 테두리 표시, *True | False
    shadow = True,  # 그림자 표시, 테두리가 있을 때만 유효, True | *False
    borderpad = 1,  # 여백
    labelspacing = 0.5  # 줄간격
)
plt.show()
```

```python
# xlim(), ylim() : 각각 X축/Y축 범위 지정하거나 반환
plt.plot([1,2,3,4], [2,3,5,10])
plt.xlim([0, 5])
plt.ylim([0, 20])
xmin, xmax = plt.xlim()  

plt.show()

print('x축 최대값 :', xmax)
```

```python
# axis() : X, Y축 범위 지정/반환
plt.plot([1,2,3,4], [2,3,5,10])
plt.axis([0,5,0,20])  # [xmin, xmax, ymin, ymax] | on | off | equal | scaled | tight | auto | normal | image | square
plt.show()
```

```python
# 선 종류 지정하기
plt.plot([1,2,3],[4,4,4], '-', label='solid')
plt.plot([1,2,3],[3,3,3], '--', label='dashed')
plt.plot([1,2,3],[2,2,2], ':', label='dotted')
plt.plot([1,2,3],[1,1,1], '-.', label='dashdot')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol = 4)
plt.show()
```

```python
# fill_between() : x축과 만나는 영역 채우기
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.fill_between(x[1:3],y[1:3], alpha=0.5)  # (x[1], y[1]), (x[2], y[2]), (x[2], 0), (x[1], 0) 을 잇는 영역이 채워짐
```

```python
# fill_betweenx() : y축과 만나는 영역 채우기
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.fill_betweenx(y[1:3],x[1:3], alpha=0.5)  # (x[1], y[1]), (x[2], y[2]), (0, y[2]), (0, y[1]) 을 잇는 영역이 채워짐
```

```python
# fill_between() : 두 그래프 사이의 영역 채우기
x = [1,2,3,4]
p = [2,3,5,10]
q = [1,2,4,8]

plt.plot(x,p)
plt.plot(x,q)
plt.fill_between(x[1:3],p[1:3], q[1:3], color='lightgray', alpha=0.5)  # (x[1], p[1]), (x[2], p[2]), (x[2], q[2]), (x[1], q[1]) 을 잇는 영역이 채워짐
```

```python
# fill() : 임의의 영역 채우기
x = [1,2,3,4]
p = [2,3,5,10]
q = [1,2,4,8]

plt.plot(x,p)
plt.plot(x,q)
plt.fill([1.9, 1.9, 3.1, 3.1], [1, 4, 6, 3], color='lightgray', alpha=0.5)  
```

```python
# xscale(), yscale() : 축 스케일 지정하기
plt.subplot(1, 2, 1)
x = np.linspace(0, 5, 100)
y = np.exp(x)
plt.plot(x, y)
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.yscale('log')  # *linear | log | symlog | logit
plt.show()
```

```python
# grid() : 그리드 설정하기
x = np.arange(0, 2, 0.2)
plt.plot(x, x**3)
plt.grid(True,  # 표시할지 여부 
         axis = 'y',  # 그리드 방향, *both | x | y
         color = 'red',  # 그리드 색상
         alpha = 0.5,  # 그리드 투명도
         linestyle = '--'  # 그리드 스타일
        )

plt.show()
```

```python
# xticks(), yticks() : 축 눈금 표시하기
# tick_params() : 축 눈금 속성 설정하기
x = np.arange(0, 2, 0.2)
plt.plot(x, x**3)
plt.xticks(x, labels=['Jan', '', 'Feb', '', 'Mar', '', 'Apr', '', 'May', ''])  # 입력한 눈금의 갯수와 레이블의 개수가 같아야 함
plt.yticks(np.arange(1,6))
plt.tick_params(axis = 'x',  # 축 방향, x | y | *both
                direction = 'in',  # 눈금 방향, in | out | *inout
                length = 10,  # 눈금 길이
                pad = 15,  # 눈금과 레이블의 거리
                labelsize = 14,  # 레이블 크기
                labelcolor = 'blue',  # 레이블 색상
                top = True,  # 눈금 표시 위치,  top/bottom/left/right,  True | False
                width = 2,  # 눈금 너비
                color = 'r'  # 눈금 색상                
               )  

plt.show()
```

```python
# title() : 제목 표시하기
x = np.arange(0, 2, 0.2)
plt.plot(x, x**3)
plt.title('Graph Title',  # 표시 문구 지정
          loc = 'left',  # 표시 위치, left | *center | right
          pad = 20,  # 타이틀과 그래프 간 간격
          fontdict = {'fontsize':16, 'fontweight':'bold'}  # 폰트 지정
         )
plt.title('Graph Title', loc='right', fontdict={'fontsize': 14, 'weight' : 'light'})  # 그런데 폰트 굵기는 변경 안되는듯

plt.show()
```

```python
# axhline() : 수평선 표시
# axvline() : 수직선 표시
# hlines() : 지정한 점을 따라 수평선 표시
# vlines() : 지정한 점을 따라 수직선 표시
x = np.arange(0, 4, 0.5)
plt.plot(x, x**2 - 4)
plt.axhline(4,  # y값, 수평선 위치
            0.1, 0.9,  # x값, 0에서 1 사이의 값 입력
            color = 'lightgray',
            linestyle = '--',
            linewidth = 2
           )
plt.hlines(-3,  # y값, 수평선 위치
           0.1, 0.9,  # x값, 좌표값 직접 입력
           color = 'darkgray',
           linestyle = 'solid',
           linewidth = 3
          )
plt.axvline(0.8,  # x값, 수직선 위치
           0.1, 0.65,  # y값, 0에서 1 사이의 값 입력
           color = 'red')

plt.vlines(2,  # x값, 수직선 위치
          -2, 4,  # y값, 좌표값 직접 입력
          color = 'crimson')
plt.show()
```

```python
# errorbar() : 에러바(오차막대) 표시하기
x = [1,2,3,4]
y1 = [3,3,3,3]
y2 = [13,13,13,13]
y3 = [23,23,23,23]

plt.errorbar(x, y1, yerr = [2, 3, 1, 2])  # plt.plot 없이 표시함, 오차는 위아래 대칭으로 나타남
plt.errorbar(x, y2, yerr = [(1,2,3,4), (4,3,2,1)])  # 비대칭 편차 나타내기
plt.errorbar(x, y3, yerr = [1,1,1,1], uplims = [True, False, True, False], lolims = [False, False, True, True])  # 상하한 기호 표시 

plt.show()
```

```python
# text() : 텍스트 삽입하기
x = np.arange(0, 2, 0.2)
plt.plot(x, x**3)
font1 = {
    'family' : 'Times New Roman',  # 폰트 종류
    'color' : 'darkgreen',  # 폰트 색상
    'weight' : 'bold',  # 폰트 굵기
    'size' : 14,  # 폰트 크기
    'alpha' : 0.7,  # 폰트 투명도
    'style' : 'italic',  # 폰트 스타일
    }
box1 = {
    'boxstyle' : 'round',  # 박스 스타일,  round | square
    'ec' : (0.5, 0.5, 1),  # edge color
    'fc' : (0.8, 0.8, 1),  # face color
    'linestyle' : '-.',  # 테두리 스타일
    'linewidth' : 3,  # 테두리 두께
    
}
plt.text(1, 0.4,  # 표시할 좌표
         'y = x^3',  # 표시할 내용, ' '로 묶어야 함   
         rotation = 35,  # 텍스트 회전
         fontdict = font1,  # 텍스트 종류 세팅
         bbox = box1,  # 텍스트 박스 세팅
        )
plt.show()
```

```python
# 수학적 표현 사용하기
x = np.arange(0, 5, 0.2)
plt.plot(x, x**3)
plt.text(0, 5, r'$\alpha > \beta$')  # 그리스 문자
plt.text(0, 15, r'$y = x_1^3$')  # 위첨자, 아래첨자
plt.text(0, 25, r'$\frac{1}{2} + \frac{3}{4} = \frac{5}{4}$')  # 분수
plt.text(0, 35, r'$\sqrt{2} + \sqrt[3]{x} = y$')  # 제곱근
plt.text(0, 45, r'$\acute a, \bar a, \tilde a$')  # 액센트
plt.text(0, 55, r'$\vec a \cdot \vec a = |\vec a|^2$')
plt.text(0, 65, r'$\overline{abc}$')
plt.text(0, 75, r'$\sin(x) \ \cos(x) \ \tan(x)$')
plt.text(0, 85, r'$\lim_{x\rightarrow 2} (x^2 - x + 2)$')
plt.text(0, 95, r'$\sum_{n=0}^{10}{(n^2 + n)}$')

plt.show()
```
