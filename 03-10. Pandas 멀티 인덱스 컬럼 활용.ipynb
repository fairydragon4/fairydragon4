### single column to multi columns

```python
# 샘플 데이터프레임 호출
PATH = 'D:/python/notebooks/'
fn = 'multicolumn_sample.xlsx'
raw = pd.read_excel(PATH + fn, index_col = '고객번호')
raw
```

```python
# 멀티 컬럼으로 변경 - 방안1
raw.columns = pd.MultiIndex.from_tuples(
    [tuple(x.split('_')) for x in raw.columns]
)
raw
```

```python
# 재호출
raw = pd.read_excel(PATH + fn, index_col = '고객번호')
```

```python
# 멀티 컬럼으로 변경 - 방안2
col1 = list()
col2 = list()
for i in raw.columns:
    col1.append(i.split('_')[0])
    col2.append(i.split('_')[1])

raw.columns = [col1, col2]
# raw.reset_index(inplace=True)
raw
```

```python
# 파일로 출력
raw.to_excel(PATH + fn.split('.')[0] + '_out.xlsx')
```

### multi column to single columns

```python
# 샘플 데이터프레임 호출-2 : 멀티 컬럼 호출하기
PATH = 'D:/python/notebooks/'
fn = 'multicolumn_sample2.xlsx'
raw = pd.read_excel(PATH + fn, header = [0, 1], index_col = 0)  # 헤더를 [ , ] 으로 묶음, 인덱스는 번호로 지정해야 됨
raw
```

```python
raw.sum(axis=1, level=1)
```

```python
# 싱글 컬럼으로 변경 
cols = list()
for a, b, in list(raw.columns):
    if  b == "":
        cols.append(a)
    else:
        cols.append(f'{a}_{b}')

raw.columns = cols
# raw.index.name = '고객번호'  # 필요시 부여

raw
```

```python
raw.to_excel(PATH + fn.split('.')[0] + '_out.xlsx')
```
